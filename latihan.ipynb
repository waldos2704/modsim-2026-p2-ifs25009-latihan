# Contoh Matplotlib

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat line chart
plt.plot(x, y)

# Memberi judul dan label
plt.title("Contoh Line Chart Sederhana")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")

# Menampilkan grafik
plt.show()
```

**Analisis**: Grafik di atas menunjukkan hubungan linear positif antara variabel pada sumbu X dan sumbu Y. Setiap kenaikan 1 satuan pada sumbu X diikuti oleh kenaikan 2 satuan pada sumbu Y, sehingga garis yang terbentuk lurus dan menanjak secara konsisten
